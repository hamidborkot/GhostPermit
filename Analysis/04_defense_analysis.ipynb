# 04 Defense Analysis

This notebook summarizes LlamaGuard and baseline defense results from `../data/defense_results.csv`.

```python
import pandas as pd
df = pd.read_csv('../data/defense_results.csv')
df.head()
```

```python
detector_col = 'detector' if 'detector' in df.columns else ('guard_model' if 'guard_model' in df.columns else ('defense_model' if 'defense_model' in df.columns else 'model'))
detected_col = 'attack_detected' if 'attack_detected' in df.columns else ('detected' if 'detected' in df.columns else ('flagged' if 'flagged' in df.columns else ('blocked' if 'blocked' in df.columns else 'is_flagged')))
summary = df.groupby(detector_col)[detected_col].agg(['count', 'sum', 'mean']).reset_index()
summary
```
