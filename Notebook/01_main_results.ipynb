# 01 Main Results

This notebook summarizes GhostPermit main results from `../data/ghostpermit_results.csv`.

```python
import pandas as pd
df = pd.read_csv('../data/ghostpermit_results.csv')
df.head()
```

```python
summary = df.groupby('model')['asr'].agg(['count', 'mean']).reset_index()
summary['mean'] = (summary['mean'] * 100).round(2)
summary
```
