# 03 AgentWorm Analysis

This notebook summarizes multi-hop propagation results from `../data/agentworm_results.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/agentworm_results.csv')
df.head()
```

```python
hop_col = 'hop' if 'hop' in df.columns else ('hops' if 'hops' in df.columns else 'step')
success_col = 'success' if 'success' in df.columns else ('infected' if 'infected' in df.columns else ('propagated' if 'propagated' in df.columns else 'asr'))
g = df.groupby(hop_col)[success_col].mean().reset_index().sort_values(hop_col)
g
```

```python
plt.plot(g[hop_col], g[success_col], marker='o')
plt.xlabel('Hop')
plt.ylabel('Propagation Success Rate')
plt.title('AgentWorm Multi-Hop Propagation')
plt.show()
```
