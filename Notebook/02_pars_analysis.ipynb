# 02 PARS Analysis

This notebook analyzes the relationship between PARS and GhostPermit ASR.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/ghostpermit_results.csv')
combo_df = df.groupby('combo').agg(mean_asr=('asr', 'mean'), pars=('pars', 'mean')).reset_index()
combo_df
```

```python
plt.scatter(combo_df['pars'], combo_df['mean_asr'])
for _, r in combo_df.iterrows():
    plt.annotate(r['combo'], (r['pars'], r['mean_asr']))
plt.xlabel('PARS')
plt.ylabel('Mean ASR')
plt.title('PARS vs Mean ASR')
plt.show()
```
